# Acute Oral Toxicity 분류 모델 개발

**데이터 출처**: NTP Integrated Chemical Environment (ICE) — Acute Oral Toxicity (Rat LD50)

**목표**: 화학물질의 2D molecular descriptor를 이용해 급성 경구독성 여부를 분류하는 머신러닝 모델 개발

**레이블 정의**: GHS 기준 LD50 ≤ 300 mg/kg → Toxic (1), LD50 > 300 mg/kg → Non-toxic (0)

---

## 전체 파이프라인

```
Step 1. 원본 데이터 전처리 (ICE xlsx → 화합물별 중앙값 LD50 계산 → GHS 이진 레이블)
Step 2. RDKit으로 2D descriptor 계산
Step 3. Descriptor 전처리 (NaN 제거, 저분산 제거, 표준화)
Step 4. SelectKBest로 최적 feature 개수(k) 탐색
Step 5. Logistic Regression / MLP / SVM 비교
Step 6. Decision Tree / Random Forest / XGBoost 비교
Step 7. 전체 6모델 결과 종합 및 시각화
```

---
# Step 1. 원본 데이터 전처리

ICE에서 받은 `acute_oral.xlsx`를 정리해서 분류 모델에 쓸 수 있는 데이터셋을 만든다.

**처리 순서**
1. Endpoint = `LD50`, Mixture = `Chemical`, SMILES 있는 행만 선택
2. 같은 화합물(CASRN)에 여러 LD50이 있을 때 → 중앙값으로 대표값 계산
3. GHS 기준으로 이진 레이블 생성: LD50 ≤ 300 mg/kg → 1 (Toxic), 나머지 → 0

````python
import pandas as pd

# 원본 파일 불러오기
df_raw = pd.read_excel('acute_oral.xlsx', sheet_name='Data')
print('원본 shape:', df_raw.shape)
print('Endpoint 종류:', df_raw['Endpoint'].value_counts().to_dict())
````

````python
# LD50 + 단일화합물(Chemical) + SMILES 있는 행만 선택
df_filtered = df_raw[
    (df_raw['Endpoint'] == 'LD50') &
    (df_raw['Mixture'] == 'Chemical') &
    (df_raw['SMILES'].notna())
].copy()

print('필터 후 shape:', df_filtered.shape)
````

````python
# 화합물별 중앙값 LD50 계산 (같은 CASRN에 여러 실험값이 있을 수 있음)
df_compound = df_filtered.groupby('CASRN').agg(
    LD50_median=('Response', 'median'),
    SMILES=('SMILES', 'first'),
    Chemical_Name=('Chemical_Name', 'first')
).reset_index()

print('중복 제거 후 고유 화합물 수:', len(df_compound))
````

````python
import matplotlib.pyplot as plt
import numpy as np

# LD50 분포 확인 (log scale)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(df_compound['LD50_median'], bins=50, color='steelblue', edgecolor='white')
axes[0].set_xlabel('LD50 (mg/kg)')
axes[0].set_ylabel('Count')
axes[0].set_title('LD50 distribution (linear)')

axes[1].hist(np.log10(df_compound['LD50_median'] + 1), bins=50, color='steelblue', edgecolor='white')
axes[1].axvline(np.log10(300), color='red', linestyle='--', label='LD50=300 (GHS cutoff)')
axes[1].set_xlabel('log10(LD50 + 1)')
axes[1].set_ylabel('Count')
axes[1].set_title('LD50 distribution (log scale)')
axes[1].legend()

plt.tight_layout()
plt.show()
````

````python
# GHS 이진 레이블 생성
# GHS Category 1-3: LD50 <= 300 mg/kg → Toxic (1)
# GHS Category 4-5: LD50 > 300 mg/kg  → Non-toxic (0)
df_compound['label'] = (df_compound['LD50_median'] <= 300).astype(int)

print('label 분포:')
print(df_compound['label'].value_counts())
print(f"\nToxic 비율: {df_compound['label'].mean():.1%}")
````

---
# Step 2. RDKit으로 2D Descriptor 계산

SMILES → 분자 구조(mol) → 217개 2D descriptor 계산

**제외 조건**
- SMILES에 `.`이 포함된 경우 → 단일 분자가 아님 (염, 혼합물 등)
- `Chem.MolFromSmiles()`가 `None`을 반환하는 경우 → 유효하지 않은 SMILES

````python
from rdkit import Chem
from rdkit.Chem import Descriptors

desc_names = [d[0] for d in Descriptors.descList]
desc_fns   = [d[1] for d in Descriptors.descList]
print('계산할 descriptor 수:', len(desc_names))
````

````python
import numpy as np

rows = []
skipped_dot   = 0
skipped_invalid = 0

for _, row in df_compound.iterrows():
    smi = row['SMILES']

    # 단일 분자 구조가 아닌 경우 제외
    if '.' in smi:
        skipped_dot += 1
        continue

    mol = Chem.MolFromSmiles(smi)
    if mol is None:
        skipped_invalid += 1
        continue

    # descriptor 계산
    vals = []
    for fn in desc_fns:
        try:
            vals.append(fn(mol))
        except:
            vals.append(np.nan)

    rows.append([
        row['CASRN'], row['Chemical_Name'], smi,
        row['LD50_median'], row['label']
    ] + vals)

cols = ['CASRN', 'Chemical_Name', 'standardized_smi', 'LD50_median', 'label'] + desc_names
df_desc = pd.DataFrame(rows, columns=cols)

print(f'제외 (점 포함): {skipped_dot}')
print(f'제외 (유효하지 않은 SMILES): {skipped_invalid}')
print(f'최종 shape: {df_desc.shape}')
print('label 분포:')
print(df_desc['label'].value_counts())
````

````python
# 저장
df_desc.to_csv('acute_oral_2Ddesc.csv', index=False)
print('저장 완료: acute_oral_2Ddesc.csv')
````

---
# Step 3. Descriptor 전처리

수업에서 Skin Irritation 데이터에 적용했던 것과 동일한 전처리 과정:
1. **NaN 열 제거** — 일부 descriptor는 특정 분자에서 계산 실패
2. **저분산 열 제거** — 표준편차 0.01 미만이면 거의 모든 화합물에서 같은 값 → 분류에 도움 안 됨
3. **표준화** — Logistic Regression / MLP / SVM은 스케일에 민감하므로 `StandardScaler` 적용

> ⚠️ 트리 기반 모델(DT, RF, XGBoost)은 표준화 불필요 — Step 6에서는 표준화 없이 사용

````python
import pandas as pd

df = pd.read_csv('acute_oral_2Ddesc.csv')
print('불러온 shape:', df.shape)
df.head()
````

````python
# y(label)와 X(descriptor) 분리
y = df['label']
X = df.drop(columns=['CASRN', 'Chemical_Name', 'standardized_smi', 'LD50_median', 'label'])

# NaN 열 제거
X = X.dropna(axis=1)
print('NaN 제거 후:', X.shape)

# 저분산 열 제거 (표준편차 0.01 미만)
X = X.loc[:, X.std() >= 0.01]
print('저분산 제거 후:', X.shape)

print('\ny 분포:')
print(y.value_counts())
````

---
# Step 4. SelectKBest로 최적 K 탐색

`SelectKBest(f_classif)`로 descriptor 상위 k개를 선택하면서 5-fold CV 정확도가 가장 높은 k를 찾는다.

- `f_classif`: 각 descriptor와 label 사이의 ANOVA F-value 계산 → 높을수록 분류에 유용
- k가 너무 작으면 정보 부족(underfitting), 너무 크면 노이즈 포함(overfitting)

````python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt

k_candidates = [5, 10, 20, 30, 40, 50, 70, 100]
cv_scores = []

for k in k_candidates:
    selector = SelectKBest(f_classif, k=k)
    X_sel = selector.fit_transform(X, y)

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_sel)

    clf = LogisticRegression(max_iter=1000, random_state=0)
    cv = cross_val_score(clf, X_scaled, y, cv=5).mean()
    cv_scores.append(cv)
    print(f'k={k:3d}  CV5={cv:.3f}')

best_k = k_candidates[cv_scores.index(max(cv_scores))]
print(f'\n최적 k: {best_k}  (CV={max(cv_scores):.3f})')
````

````python
# k vs CV 그래프
plt.figure(figsize=(8, 4))
plt.plot(k_candidates, cv_scores, marker='o', color='steelblue')
plt.axvline(best_k, color='red', linestyle='--', label=f'best k={best_k}')
plt.xlabel('k (number of descriptors)')
plt.ylabel('5-fold CV accuracy')
plt.title('SelectKBest: k vs CV accuracy (Logistic Regression)')
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# best_k로 descriptor 선택
selector = SelectKBest(f_classif, k=best_k)
selector.fit(X, y)
selected_cols = X.columns[selector.get_support()]
X_sel = X[selected_cols]

print('선택된 descriptor:')
print(list(selected_cols))
````

---
# Step 5. 선형/신경망 모델 비교 (Logistic Regression / MLP / SVM)

같은 k의 descriptor로 3가지 모델을 학습하고 grid search로 최적 hyperparameter를 찾는다.

- **Logistic Regression**: 정규화 강도 `C` 탐색
- **MLP**: hidden layer 크기 탐색
- **SVM**: `C` × `kernel` 조합 탐색

````python
from sklearn.preprocessing import StandardScaler

# 트리 제외 모델은 표준화 필요
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_sel)
````

````python
# Logistic Regression grid search
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

C_values = [0.001, 0.01, 0.1, 1, 10]

for C in C_values:
    clf = LogisticRegression(C=C, max_iter=1000, random_state=0)
    train = clf.fit(X_scaled, y).score(X_scaled, y)
    cv    = cross_val_score(clf, X_scaled, y, cv=5).mean()
    print(f'C={C:.3f}  train={train:.3f}  CV5={cv:.3f}')
````

````python
# MLP grid search
from sklearn.neural_network import MLPClassifier

hidden_sizes = [(50,), (100,), (100, 50), (200, 100)]

for hs in hidden_sizes:
    clf = MLPClassifier(hidden_layer_sizes=hs, max_iter=500, random_state=0)
    train = clf.fit(X_scaled, y).score(X_scaled, y)
    cv    = cross_val_score(clf, X_scaled, y, cv=5).mean()
    print(f'hidden={str(hs):15s}  train={train:.3f}  CV5={cv:.3f}')
````

````python
# SVM grid search
from sklearn.svm import SVC

C_values  = [0.1, 1, 10]
kernels   = ['linear', 'rbf']

for C in C_values:
    for kernel in kernels:
        clf = SVC(C=C, kernel=kernel, random_state=0)
        train = clf.fit(X_scaled, y).score(X_scaled, y)
        cv    = cross_val_score(clf, X_scaled, y, cv=5).mean()
        print(f'C={C:4.1f}  kernel={kernel:6s}  train={train:.3f}  CV5={cv:.3f}')
````

````python
import warnings
warnings.filterwarnings('ignore')

# 전체 k × 모델 실험 (Step 7 비교용)
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC

results = []
k_candidates_all = [5, 10, 20, 30, 40, 50, 70, 100]

model_configs = {
    'LogReg': LogisticRegression(C=0.1, max_iter=1000, random_state=0),
    'MLP':    MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=0),
    'SVM':    SVC(C=1, kernel='rbf', random_state=0),
}

for k in k_candidates_all:
    selector = SelectKBest(f_classif, k=k)
    X_k = selector.fit_transform(X, y)
    X_k_scaled = StandardScaler().fit_transform(X_k)

    for name, clf in model_configs.items():
        train = clf.fit(X_k_scaled, y).score(X_k_scaled, y)
        cv    = cross_val_score(clf, X_k_scaled, y, cv=5).mean()
        results.append({'model': name, 'k': k, 'train': round(train, 3), 'cv': round(cv, 3)})

results_df_linear = pd.DataFrame(results)
results_df_linear.to_csv('experiment_results_linear.csv', index=False)

print('전체 Top 5 (CV 기준):')
print(results_df_linear.sort_values('cv', ascending=False).head(5).to_string(index=False))
print()
print('모델별 베스트:')
best_linear = (results_df_linear
               .sort_values('cv', ascending=False)
               .groupby('model')
               .head(1)
               .sort_values('model'))
print(best_linear.to_string(index=False))
````

---
# Step 6. 트리 기반 모델 비교 (Decision Tree / Random Forest / XGBoost)

> 🚨 트리 기반 모델은 표준화(StandardScaler) 불필요
> 트리가 분기할 때 보는 건 "값이 어느 임계값보다 큰가/작은가?" — 절대값이 아닌 순서(order)만 본다.

````python
# Decision Tree grid search
from sklearn.tree import DecisionTreeClassifier

max_depths   = [2, 3, 5, None]
min_splits   = [2, 4, 6]

for d in max_depths:
    for s in min_splits:
        clf   = DecisionTreeClassifier(max_depth=d, min_samples_split=s, random_state=0)
        train = clf.fit(X_sel, y).score(X_sel, y)
        cv    = cross_val_score(clf, X_sel, y, cv=5).mean()
        print(f'max_depth={str(d):4s}  min_split={s}  train={train:.3f}  CV5={cv:.3f}')
````

````python
# Random Forest grid search
from sklearn.ensemble import RandomForestClassifier

n_estimators_list = [50, 100, 200]
max_depths_rf     = [3, 5, 7, None]

for n in n_estimators_list:
    for d in max_depths_rf:
        clf   = RandomForestClassifier(n_estimators=n, max_depth=d, random_state=0)
        train = clf.fit(X_sel, y).score(X_sel, y)
        cv    = cross_val_score(clf, X_sel, y, cv=5).mean()
        print(f'n={n:3d}  max_depth={str(d):4s}  train={train:.3f}  CV5={cv:.3f}')
````

````python
# XGBoost grid search
from xgboost import XGBClassifier

n_estimators_list = [50, 100, 200]
max_depths_xgb    = [3, 5, 7]
learning_rates    = [0.05, 0.1, 0.3]

for n in n_estimators_list:
    for d in max_depths_xgb:
        for lr in learning_rates:
            clf   = XGBClassifier(n_estimators=n, max_depth=d, learning_rate=lr,
                                  random_state=0, eval_metric='logloss', verbosity=0)
            train = clf.fit(X_sel, y).score(X_sel, y)
            cv    = cross_val_score(clf, X_sel, y, cv=5).mean()
            print(f'n={n:3d}  depth={d}  lr={lr:.2f}  train={train:.3f}  CV5={cv:.3f}')
````

````python
# 전체 k × 트리 모델 실험
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

results_tree = []

tree_configs = {
    'DT':  DecisionTreeClassifier(max_depth=5, random_state=0),
    'RF':  RandomForestClassifier(n_estimators=100, max_depth=5, random_state=0),
    'XGB': XGBClassifier(n_estimators=100, max_depth=5, learning_rate=0.1,
                         random_state=0, eval_metric='logloss', verbosity=0),
}

for k in k_candidates_all:
    selector = SelectKBest(f_classif, k=k)
    X_k = selector.fit_transform(X, y)

    for name, clf in tree_configs.items():
        train = clf.fit(X_k, y).score(X_k, y)
        cv    = cross_val_score(clf, X_k, y, cv=5).mean()
        results_tree.append({'model': name, 'k': k, 'train': round(train, 3), 'cv': round(cv, 3)})

results_df_tree = pd.DataFrame(results_tree)
results_df_tree.to_csv('experiment_results_tree.csv', index=False)

print('트리 모델 Top 5 (CV 기준):')
print(results_df_tree.sort_values('cv', ascending=False).head(5).to_string(index=False))
print()
print('트리 모델별 베스트:')
best_tree = (results_df_tree
             .sort_values('cv', ascending=False)
             .groupby('model')
             .head(1)
             .sort_values('model'))
print(best_tree.to_string(index=False))
````

---
# Step 7. 전체 6모델 결과 종합 및 시각화

선형 모델(LogReg, MLP, SVM)과 트리 모델(DT, RF, XGBoost)의 최고 성능을 비교한다.

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 두 결과 합치기
df_all = pd.concat([results_df_linear, results_df_tree], ignore_index=True)

# 모델별 베스트
best_all = (df_all
            .sort_values('cv', ascending=False)
            .groupby('model')
            .head(1)
            .sort_values('cv', ascending=False))

print('전체 6모델 베스트 성능:')
print(best_all.to_string(index=False))
````

````python
# 그래프 1 — 모델별 train vs CV 막대 비교
models = best_all['model'].tolist()
train_scores = best_all['train'].tolist()
cv_scores_best = best_all['cv'].tolist()

x = np.arange(len(models))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 5))
bars1 = ax.bar(x - width/2, train_scores, width, label='Train accuracy', color='steelblue', alpha=0.8)
bars2 = ax.bar(x + width/2, cv_scores_best, width, label='5-fold CV accuracy', color='coral', alpha=0.8)

ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_ylabel('Accuracy')
ax.set_title('Model comparison: Train vs CV accuracy')
ax.set_ylim(0.5, 1.05)
ax.legend()
ax.axhline(0.78, color='gray', linestyle=':', linewidth=0.8)  # baseline 참고선

for bar in bars1:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
            f'{bar.get_height():.3f}', ha='center', va='bottom', fontsize=9)
for bar in bars2:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
            f'{bar.get_height():.3f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
````

````python
# 그래프 2 — 6모델의 k vs CV 학습 곡선
fig, ax = plt.subplots(figsize=(10, 5))

colors = {'LogReg': 'steelblue', 'MLP': 'orange', 'SVM': 'green',
          'DT': 'red', 'RF': 'purple', 'XGB': 'brown'}

for model in df_all['model'].unique():
    df_m = df_all[df_all['model'] == model]
    df_m_best = (df_m
                 .sort_values('cv', ascending=False)
                 .groupby('k')
                 .head(1)
                 .sort_values('k'))
    ax.plot(df_m_best['k'], df_m_best['cv'],
            marker='o', label=model, color=colors.get(model, 'gray'))

ax.set_xlabel('k (number of descriptors)')
ax.set_ylabel('5-fold CV accuracy')
ax.set_title('Learning curve: k vs CV accuracy (all 6 models)')
ax.legend()
plt.tight_layout()
plt.show()
````

````python
# 최종 모델 선택 및 Feature Importance 시각화 (Random Forest 기준)
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest, f_classif

# best_all에서 RF의 최적 k 확인
best_rf_k = int(best_all[best_all['model'] == 'RF']['k'].values[0])
print(f'RF 최적 k: {best_rf_k}')

selector_final = SelectKBest(f_classif, k=best_rf_k)
X_final = selector_final.fit_transform(X, y)
final_cols = X.columns[selector_final.get_support()]

rf_final = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=0)
rf_final.fit(X_final, y)

# Feature importance
importances = rf_final.feature_importances_
feat_imp = pd.Series(importances, index=final_cols).sort_values(ascending=False)

plt.figure(figsize=(10, 5))
feat_imp.head(15).plot(kind='bar', color='steelblue', edgecolor='white')
plt.title('Random Forest: Top 15 feature importances')
plt.ylabel('Importance')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print('\nTop 10 important descriptors:')
print(feat_imp.head(10).to_string())
````

- Train과 CV 차이가 크면 → overfitting 의심
- 트리 단독 vs 앙상블 성능 차이 → 앙상블의 효과 확인
- Feature importance 상위 descriptor → 어떤 분자 특성이 독성과 연관 있는지 해석
